##Question 1
What age group is most affected by heart disease

```python
age = heart_df['age'].unique()
age
```

```python
age = heart_df['age'].describe()
age
```

```python
# Create age bins
bins = [20, 30, 40, 50, 60, 70, 80]
labels = ['20-29', '30-39', '40-49', '50-59', '60-69', '70-79']

heart_df['age_group'] = pd.cut(heart_df['age'], bins=bins, labels=labels, right=False)
```

```python
age_disease_counts = heart_df[heart_df['target'] == 1]['age_group'].value_counts().sort_index()
age_disease_counts
```

```python
## Plot the age disease count
age_disease_counts.plot(kind='bar', figsize=(8,5))
plt.title('Heart Disease Cases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Cases')
plt.show()
```

```python
## Percentage of cases in each age group

age_disease_percentages = (age_disease_counts / age_disease_counts.sum()) * 100
age_disease_percentages
```

The age group most affected by heart disease is the aged 50–59. These Adults  account for the largest share of heart disease cases

# Question 2
## In what age group is heart disease most prevalent


```python
age_group_counts = heart_df['age_group'].value_counts().sort_index()
age_group_counts
```

```python
## Create prevalance tabble
prevalence_table = pd.DataFrame({
    'total_people': age_group_counts,
    'heart_disease_cases': age_disease_counts
})

## Calculate prevalence (%)
prevalence_table['prevalence_%'] = (
    prevalence_table['heart_disease_cases'] /
    prevalence_table['total_people']
) * 100

## Flag reliable estimates
prevalence_table['reliable'] = prevalence_table['total_people'] >= 30

prevalence_table
```

The highest reliable heart disease prevalence is observed among adults aged 40–49 and 50–59. Estimates for younger age groups 20-29, 30-39 and the older age grouop 70-79 should be interpreted with caution due to small sample sizes

N/B : The total burden is the group 50 -59, however,
the most vulnerable is the group 40 - 49

## Question 3
What gender is more at risk of a heart disease

```python
## Groupby sex and find the mean of target
gender_risk = heart_df.groupby('sex')['target'].mean()
gender_risk
```

```python
## Plot a graph showing risk of heart disease by gender
gender_risk.plot(kind='bar')
plt.ylabel('Proportion with Heart Disease')
plt.title('Risk of Heart Disease by Gender')
plt.show()
```

The chart shows that female are more at risk of a heart disease.
